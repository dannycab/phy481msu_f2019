```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('seaborn-colorblind')
```

```python
x = np.arange(-5,5.5,0.5)
v = np.arange(-5,5.5,0.5)

X, V = np.meshgrid(x,v)

xdot = V
vdot = -X

fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(1, 2, 1)
plt.quiver(X, V, xdot, vdot)
plt.xlabel('x');
plt.ylabel('v');
plt.axis([-5,5,-5,5]);

fig.add_subplot(1, 2, 2)
plt.streamplot(X, V, xdot, vdot)
plt.xlabel('x');
plt.ylabel('v');
plt.axis([-5,5,-5,5]);
```

```python
x = np.arange(-5,5.5,0.5)
v = np.arange(-5,5.5,0.5)

X, V = np.meshgrid(x,v)

xdot = V
vdot = -np.sin(X)

fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(1, 2, 1)
plt.quiver(X, V, xdot, vdot)
plt.xlabel('x');
plt.ylabel('v');
plt.axis([-5,5,-5,5]);

fig.add_subplot(1, 2, 2)
plt.streamplot(X, V, xdot, vdot)
plt.xlabel('x');
plt.ylabel('v');
plt.axis([-5,5,-5,5]);
```

```python
from scipy.integrate import odeint

def pend(y, t, b, c):
    
    theta, omega = y
    dydt = [omega, - c*np.sin(theta)]
    
    return dydt

def pendApprox(y, t, b, c):
    
    theta, omega = y
    dydt = [omega, - c*theta]
    
    return dydt

# initial condition
b = 0.2
c = 1

y0 = [0, 0.1]

# time points
t = np.linspace(0, 10, 1001)

# solve ODE
solFull = odeint(pend, y0, t, args=(b, c))
solApprox = odeint(pendApprox, y0, t, args=(b, c))

# plot results
fig = plt.figure(figsize=(9, 9))
plt.plot(t, solFull[:, 0], label='Exact')
plt.plot(t, solApprox[:, 0], label='Approx')
plt.legend()
plt.xlabel('time')
plt.ylabel('theta(t)')
```

```python
x = np.arange(-5,5.5,0.5)
v = np.arange(-5,5.5,0.5)

X, V = np.meshgrid(x,v)

xdot = V
vdot = -0.2*V-X

xdot1 = V
vdot1 = -0.2*V-np.sin(X)

fig = plt.figure(figsize=(16, 16))
ax = fig.add_subplot(2, 2, 1)
plt.quiver(X, V, xdot, vdot)
plt.xlabel('x');
plt.ylabel('v');
plt.axis([-5,5,-5,5]);

fig.add_subplot(2, 2, 2)
plt.streamplot(X, V, xdot, vdot)
plt.xlabel('x');
plt.ylabel('v');
plt.axis([-5,5,-5,5]);

ax = fig.add_subplot(2, 2, 3)
plt.quiver(X, V, xdot1, vdot1)
plt.xlabel('x');
plt.ylabel('v');
plt.axis([-5,5,-5,5]);

fig.add_subplot(2, 2, 4)
plt.streamplot(X, V, xdot1, vdot1, density=2)
plt.xlabel('x');
plt.ylabel('v');
plt.axis([-5,5,-5,5]);
```

```python
x = np.arange(-5,10.5,0.5)
v = np.arange(-5,10.5,0.5)

X, V = np.meshgrid(x,v)
b=0.2

xdot = V
vdot = -b*V-X

xdot1 = V
vdot1 = -b*V-np.sin(X)

start = np.array([[0,0], [3,3]])
fig = plt.figure(figsize=(16, 8))
fig.add_subplot(1, 2, 1)
plt.streamplot(X, V, xdot, vdot)
plt.streamplot(X, V, xdot, vdot, start_points=start.T)
plt.plot(start[0], start[1], 'kx')

plt.xlabel('x');
plt.ylabel('v');
plt.axis([-5,5,-5,5]);

fig.add_subplot(1, 2, 2)
plt.streamplot(X, V, xdot1, vdot1)
plt.streamplot(X, V, xdot1, vdot1, start_points=start.T)
plt.plot(start[0], start[1], 'kx')
plt.xlabel('x');
plt.ylabel('v');
plt.axis([-5,10,-5,5]);
```

```python
from scipy.integrate import odeint

def pendExact(y,t,b,c):
    theta, omega = y
    dydt = [omega, -b*omega - c*np.sin(theta)]
    return dydt

def pendApprox(y,t,b,c):
    theta, omega = y
    dydt = [omega, -b*omega - c*theta]
    return dydt

b = 0.2
c = 1

y0 = [0,3]

t = np.linspace(0,100,1001)

solExact = odeint(pendExact, y0, t, args=(b,c))
solApprox = odeint(pendApprox, y0, t, args=(b,c))

fig = plt.figure(figsize=(16, 8))
plt.plot(t, solExact[:, 0], label='Exact')
plt.plot(t, solApprox[:, 0], label='Approximate')
plt.legend(loc='best')
plt.xlabel('t')
plt.ylabel('theta(t)')
```
