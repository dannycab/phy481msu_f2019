```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
def f(x):
    return x**4-2*x+1
```

```python
def trap(a,b,N):
    
    s = 0
    h = (b-a)/N
    
    s = 0.5*f(a) + 0.5*f(b)

    for k in range(1,N):
        
        s+=f(a+k*h)
        
    return h*s
```

```python
a,b = 0,2
steps = range(2,1000)
S = []
for step in steps:
      S.append(trap(a,b,step))  
```

```python
plt.plot(steps[900:],S[900:])
```
